```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 

framingham = pd.read_csv('framingham.csv')
framingham
```

```python
non_smoker = framingham.loc[framingham['currentSmoker'] == 0]
non_smoker

smoker = framingham.loc[framingham['currentSmoker'] == 1]
smoker
```

```python
group_non_smokers = non_smoker.groupby('age')['heartRate'].mean().reset_index()
group_non_smokers

group_smokers = smoker.groupby('age')['heartRate'].mean().reset_index()
group_smokers
```
